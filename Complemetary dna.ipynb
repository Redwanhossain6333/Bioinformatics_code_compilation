```python
# Read the DNA strand from a text file
with open("F:\Download Folder\dataset_30273_2(1).txt", 'r') as file:
    dna_strand = file.read().strip()

# Function to generate the complementary DNA strand
def complementary_strand(dna_strand):
    complement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    complementary = [complement[base] for base in dna_strand]
    return ''.join(complementary)

# Generate and print the complementary strand
complementary_dna = complementary_strand(dna_strand)
complementary_dna = complementary_dna[::-1]  # Reverse the complementary DNA strand
print("Original DNA Strand:", dna_strand)
print("Complementary DNA Strand:", complementary_dna)
```
